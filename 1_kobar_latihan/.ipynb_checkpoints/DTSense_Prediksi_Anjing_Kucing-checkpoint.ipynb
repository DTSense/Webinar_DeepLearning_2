# Klasifikasi Gambar Anjing vs Kucing

Pada tutorial kali ini kita akan membahas bagaimana mengelompokkan gambar menjadi gambar kucing atau gambar anjing. Kita akan membuat pengklasifikasi gambar menggunakan model `tf.keras.Sequential`.

# Importing packages

```python
import tensorflow as tf
```

```python
import matplotlib.pylab as plt
import numpy as np

import tensorflow_hub as hub
import tensorflow_datasets as tfds

from tensorflow.keras import layers
```

```python
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)
```

```python
(train_examples, validation_examples), info = tfds.load(
    'cats_vs_dogs', 
    with_info=True, 
    as_supervised=True, 
    split=['train[:80%]', 'train[80%:]'],
)

num_examples = info.splits['train'].num_examples
num_classes = info.features['label'].num_classes

print(num_examples)
print(num_classes)
```

Kita bisa lihat di bawah ini gambar pada dataset Dogs vs Cats tidak mempunyai size yang sama.

```python
for i, example_image in enumerate(train_examples.take(3)):
  print("Image {} shape: {}".format(i+1, example_image[0].shape))
```

Jadi kita perlu reformat semua gambar ke resolusi (244, 244) untuk membuat standarisasi karena kita juga akan mencoba menggunakan Transfer Learning nantinya menggunakan MobileNet.

```python
IMAGE_RES = 224
```

```python
def format_image(image, label):
  image = tf.image.resize(image, (IMAGE_RES, IMAGE_RES))/255.0
  return image, label

BATCH_SIZE = 32

train_batches      = train_examples.shuffle(num_examples//4).map(format_image).batch(BATCH_SIZE).prefetch(1)
validation_batches = validation_examples.map(format_image).batch(BATCH_SIZE).prefetch(1)
```

# Visualizing Training images

```python
image_batch, label_batch = next(iter(train_batches.take(1)))
```

Fungsi `next` untuk mengembalikan/return satu batch dari dataset. Satu batch terdiri dari tuple of (*many images*, *many labels*).

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
```

```python
plotImages(image_batch[:5]) 
```

# Build Model

<img src="https://i.imgur.com/ebkMGGu.jpg" width="100%"/>

Model yang akan kita buat terdiri dari 4 block layer. 3 block untuk mengekstrak fitur pada gambar dan block layer terakhir digunakan untuk classifier. Pada Feature Extractors kita menggunakan convolutional layer, BatchNorm, Pooling dan DropOut. Layer terakhir adalah Fully connected layer dengan 512 units dan relu activation function.

* **Input Layer **: Layer yang mewakili input untuk data gambar. Ini akan membentuk / reshape gambar menjadi array satu dimensi. Sebagai contoh jika dimensi gambar adalah 64x64 = 4096, maka akan diubah menjadi (4096,1) array.
* **Conv Layer **: Layer yang digunakan untuk mengekstrak fitur dari gambar.
* **Pooling Layer **: Layer yang digunakan untuk  mengurangi volume spasial gambar setelah proses konvolusi.
* **Fully Connected Layer **: Layer yang menghubungkan satu layer ke layer yang lain
* **Output Layer **: Ini adalah layer terakhir yang memberikan nilai yang diprediksi sebagai output.

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3,3), activation='relu', input_shape=(IMAGE_RES, IMAGE_RES, 3)),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Dropout(0.25),

    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.25),
    
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Dropout(0.25),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(2)
])

model.summary()
```

### Compile the model

Seperti biasa, kita menggunakan `adam` optimizer. Di karenakan kita akan mengeluarkan output dengan kategori softmax, maka kita menggunakan `sparse_categorical_crossentropy` sebagai loss function.

```python
model.compile(
  optimizer='adam',
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])

EPOCHS = 6
history = model.fit(train_batches,
                    epochs=EPOCHS,
                    validation_data=validation_batches)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# class_names = np.array(info.features['label'].names)
class_names = np.array(['Kochenk', 'Anjhenk'])
class_names
```

```python

image_batch = image_batch.numpy()
label_batch = label_batch.numpy()
```

```python
predicted_batch = model.predict(image_batch)
predicted_batch = tf.squeeze(predicted_batch).numpy()
predicted_ids = np.argmax(predicted_batch, axis=-1)
predicted_class_names = class_names[predicted_ids]
predicted_class_names
```

```python
print("Labels: ", label_batch)
print("Predicted labels: ", predicted_ids)
```

```python
plt.figure(figsize=(10,9))
for n in range(30):
  plt.subplot(6,5,n+1)
  plt.subplots_adjust(hspace = 0.3)
  plt.imshow(image_batch[n])
  color = "blue" if predicted_ids[n] == label_batch[n] else "red"
  plt.title(predicted_class_names[n].title(), color=color)
  plt.axis('off')
_ = plt.suptitle("Model predictions (blue: correct, red: incorrect)")
```

## Download Classifier

Unduh model MobileNet dan buat ML model dari MobileNet.
MobileNet require gambar 224 $ \ kali $ 224 piksel, dalam 3 channel warna (RGB).

```python
import tensorflow as tf

import matplotlib.pylab as plt

import tensorflow_hub as hub
import tensorflow_datasets as tfds

from tensorflow.keras import layers

IMAGE_RES = 224
```

```python
CLASSIFIER_URL ="https://tfhub.dev/google/tf2-preview/mobilenet_v2/classification/2"

model = tf.keras.Sequential([
    hub.KerasLayer(CLASSIFIER_URL, input_shape=(IMAGE_RES, IMAGE_RES, 3))
])
```

```python
import numpy as np
import PIL.Image as Image
import os
```

```python
os.listdir()
```

```python
test_gambar = tf.keras.utils.get_file('image_6.jpg','https://i.pinimg.com/originals/0f/66/f9/0f66f941b5ec2b2df95710e012416ddc.jpg', cache_dir='sample_data')
test_gambar = Image.open(test_gambar).resize((IMAGE_RES, IMAGE_RES))
test_gambar 
```

```python
test_gambar = np.array(test_gambar)/255.0
test_gambar.shape
```

```python
result = model.predict(test_gambar[np.newaxis, ...])
result.shape
```

Hasilnya adalah 1001 elemen vektor logits, memberi peringkat probabilitas setiap kelas untuk gambar tersebut.

Jadi ID kelas atas dapat ditemukan dengan argmax, dimana kita bisa mendapatkan index yang mempunyai probabilitas tertinggi. 

Untuk mengetahui klasifikasi apa sebenarnya dari gambar tersebut kita perlu mendownload dataset label dari ImageNetLabels.

```python
predicted_class = np.argmax(result[0], axis=-1)
predicted_class
```

## Decode the Prediction

Untuk melihat predikat_kelas di dataset ImageNet, unduh label ImageNet dan subset baris yang diprediksi oleh model.

```python
labels_path = tf.keras.utils.get_file('ImageNetLabels.txt','https://storage.googleapis.com/download.tensorflow.org/data/ImageNetLabels.txt')
imagenet_labels = np.array(open(labels_path).read().splitlines())

plt.imshow(test_gambar)
plt.axis('off')
predicted_class_name = imagenet_labels[predicted_class]
_ = plt.title("Prediction: " + predicted_class_name.title())
```

Sekarang Model kita bisa memprediksi dengan benar

```python
labels_path = tf.keras.utils.get_file('ImageNetLabels.txt','https://storage.googleapis.com/download.tensorflow.org/data/ImageNetLabels.txt')
imagenet_labels = np.array(open(labels_path).read().splitlines())
```

```python
image_batch, label_batch = next(iter(train_batches.take(1)))
image_batch = image_batch.numpy()
label_batch = label_batch.numpy()

result_batch = model.predict(image_batch)

predicted_class_names = imagenet_labels[np.argmax(result_batch, axis=-1)]
predicted_class_names
```

```python
plt.figure(figsize=(10,9))
for n in range(30):
  plt.subplot(6,5,n+1)
  plt.subplots_adjust(hspace = 0.3)
  plt.imshow(image_batch[n])
  plt.title(predicted_class_names[n])
  plt.axis('off')
_ = plt.suptitle("ImageNet predictions")
```

```python
URL = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/2"
feature_extractor = hub.KerasLayer(URL,
                                   input_shape=(IMAGE_RES, IMAGE_RES,3))
```

```python
feature_batch = feature_extractor(image_batch)
print(feature_batch.shape)
```

```python
feature_extractor.trainable = False
```

```python
model = tf.keras.Sequential([
  feature_extractor,
  layers.Dense(2)
])
```

```python
model.compile(
  optimizer='adam',
  loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])

EPOCHS = 6
history = model.fit(train_batches,
                    epochs=EPOCHS,
                    validation_data=validation_batches)
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(EPOCHS)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```
